```python
iris = load_iris()
```

iris数据集 有150个鸢尾花对象，每个对象有4个特征   
'data' array数组 大小为150×4  
'feature_names' 列表 4个特征的名字，分别是 萼片长宽 花瓣长宽  
'target_names' 分别是3个不同种类的名字  
'target' 是150个鸢尾花对象的类别(0,1,2)  
鸢尾花详细分析https://www.jianshu.com/p/52b86c774b0b

# make_classfication

```python
X1, y1 = make_classification()
```

sklearn.datasets.make_classification  
(n_samples=100,   样本数   
n_features=20,    特征数   
n_informative=2,   有效特征数   
n_redundant=2,    冗余特征数   
n_repeated=0,    重复特征数   
n_classes=2,    类别数   
n_clusters_per_class=2,  簇的个数   
weights=None,     权重数组 分配给每个类别的样本的比例，默认是平衡的   
flip_y=0.01,       
class_sep=1.0,   
hypercube=True,  
shift=0.0,  
scale=1.0,  
shuffle=True, 
random_state=None)  随机数种子 
 
详见官网https://scikit-learn.org/stable/modules/generated/sklearn.datasets.make_classification.html#sklearn.datasets.make_classification
博客https://www.cnblogs.com/xiaoyuanqujing/articles/11656837.html

```python
plt.figure(figsize=(10, 10))

plt.subplot(221)
plt.title("One informative feature, one cluster per class", fontsize=12)
X1, y1 = make_classification(n_samples=1000, random_state=1, n_features=2, n_redundant=0, n_informative=1,
                                      n_clusters_per_class=1)
plt.scatter(X1[:, 0], X1[:, 1], marker='*', c=y1)

plt.subplot(222)
plt.title("Two informative features, one cluster per class", fontsize=12)
X1, y1 = make_classification(n_samples=1000, random_state=1, n_features=2, n_redundant=0, n_informative=2,
                                      n_clusters_per_class=1)
plt.scatter(X1[:, 0], X1[:, 1], marker='*', c=y1)

plt.subplot(223)
plt.title("Two informative features, two clusters per class", fontsize=12)
X1, y1 = make_classification(
    n_samples=1000, random_state=1, n_features=2, n_redundant=0, n_informative=2)
plt.scatter(X1[:, 0], X1[:, 1], marker='*', c=y1)


plt.subplot(224)
plt.title("Multi-class, two informative features, one cluster",
          fontsize=12)
X1, y1 = make_classification(n_samples=1000, random_state=1, n_features=2, n_redundant=0, n_informative=2,
                                      n_clusters_per_class=1, n_classes=4)
plt.scatter(X1[:, 0], X1[:, 1], marker='*', c=y1)
plt.show()
```
